```
from pandas import read_csv as rcsv
import numpy as np
import matplotlib.pyplot as plt
```

```
'''(1) LECTURA DE LOS DATOS Y ALMACENAMIENTO EN ARRAY DE NUMPY'''

# Pra cargar los datos
# data = rcsv('/content/drive/My Drive/Proyecto 02/datos.csv') # Compu Carlos
data = rcsv('/content/drive/My Drive/Classroom/Machine Learning C3 - 2 GRU/Ricardo Andres Del Rio Guzman - datos.csv') # Compu Ricky

# Creación del array con los datos:
data_array = np.array(data.values)
data_array
```

```
'''(2) CREACIÓN DE UNA FUNCIÓN QUE PERMITA CALCULAR MEDIDAS DESCRITIVAS Y GRAFICAR:'''


# for i in range(1,7):
#   array_dev_rel = array_dev_est / array_aritmetico
# array_dev_rel

# def calculo_limpieza(array):
#   funcs = [np.max, np.min, np.mean, np.std]
  
#   medidas_descriptivas = np.zeros((6,5))
#   for i in range(1,7):
#     for i in range(0,6):
#       medidas_descriptivas[:,0][i-1] = np.max(data_array[:,i])
#       medidas_descriptivas[:,1][i-1] = data_array[:,i].min()
#   return medidas_descriptivas

# print(calculo_limpieza(data_array))
  

'''(2.a.1) CALCULO DE LOS MÁXIMOS DE CADA COLUMNA'''
array_maximos = np.zeros(6)
for i in range(1,7):
  maxi = data_array[:,i].max()
  array_maximos[i-1] = maxi
array_maximos
```

```
'''(2.a.2) CALCULO DE LOS MÍNIMOS DE CADA COLUMNA'''
array_minimos = np.zeros(6)

for n in range(1,7):
  mini = data_array[:,n].min()
  array_minimos[n-1]=mini
array_minimos
```

```
'''(2.a.3) CALCULO DE LOS PRMEDIOS (MEDIA ARITMÉTICA) DE CADA COLUMNA'''
array_aritmetico = np.zeros(6)
for k in range(1,7):
  media = data_array[:,k].mean()
  array_aritmetico[k-1]=media
array_aritmetico
```

```
'''(2.a.4) CALCULO DE LAS DESVIACIONES ESTÁNDAR DE CADA COLUMNA'''
array_dev_est = np.zeros(6)
for i in range(1,7):
  array_dev_est[i-1] = data_array[:,i].std()
  print(np.std(data_array[:,i]))
array_dev_est
```

```
'''(2.a.5) CALCULO LAS DESVIACIONES ESTÁNDAR RELATIVA DE CADA COLUMNA'''
for i in range(1,7):
  array_dev_rel = array_dev_est / array_aritmetico
array_dev_rel
```

```
'''(2.b.1) GRÁFICO DE TEMPERATURAS MÁXIMAS'''
plt.plot(data_array[:,0],data_array[:,1])
plt.xlabel('Dias')
plt.ylabel('Grados')
plt.title('Temperatura Maxima')
plt.show()

# En el siguiente gráfico podemos ver que la temperatura oscila, lo que 
# interpretamos puede deberse a las estaciones del año. Se observan días en los
# que la temperatura máxima se dispara a temperaturas impensables, lo que 
# seguramente se debe a errores de medición.
```

```
'''(2.c.1) CREACIÓN DE GRÁFICOS BOXPLOT
Para a visualización de Outliers'''
fig, axs = plt.subplots(2, 3)
axs[0,0].boxplot(data_array[:,1])
axs[0,1].boxplot(data_array[:,2])
axs[0,2].boxplot(data_array[:,3])
axs[1,0].boxplot(data_array[:,4])
axs[1,1].boxplot(data_array[:,5])
axs[1,2].boxplot(data_array[:,6])
fig.set_size_inches(8, 8)
fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                    hspace=0.4, wspace=0.3)
```

```
'''(2.c.2) ELIMINACIÓN DE OUTLIERS'''

def moda(columna):
  # Esta función calcula la moda de una columna dada.
  (values,counts) = np.unique(data_array[:,columna],return_counts=True)
  ind = np.argmax(counts) # Retorna el índice del valor más grande del vector counts.
  return values[ind] # Se retorna el valor correspondiente del vector values.
```

```
# Modificación de los datos, eliminando los valores muy alejados
for i in range(1,7):
  indices_arriba = np.where(data_array[:,i] > (array_aritmetico[i-1] + 3*array_dev_est[i-1]))
  for k in indices_arriba:
    data_array[:,i][k] = moda(i)
  indices_abajo = np.where(data_array[:,i] < (array_aritmetico[i-1] - 3*array_dev_est[i-1]))
  for h in indices_abajo:
    data_array[:,i][h] = moda(i)
    
```

```
'''(2.d) GRÁFICOS CON LOS DATOS MODIFICADOS'''

fig, axs = plt.subplots(3, 2)

# Se muestran los gráficos con los datos modificados:
for i in range(1,7):
  #plt.plot(data_array[:,0],data_array[:,i])
  axs[i + ((4**int(i>3))*-1), int(i>3)].plot(data_array[:,0],data_array[:,i])

fig.set_size_inches(20, 15)
fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                    hspace=0.4, wspace=0.3)
plt.show()
```

```
 '''(3) CREACION DE LAS COLUMNAS rs y rsacum'''
# Creación de columna rs:
if len(data_array[1]) == 7:
  data_array = np.c_[data_array, data_array[:,5]*data_array[:,6]]
  
  # Creación de columna rsacum:
if len(data_array[1]) == 8:
  data_array = np.c_[data_array, np.cumsum(data_array[:,7])]
print('La forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))
```

```
'''(4) CREACION DE COLUMNA CON EL PORCENTAJE QUE T2 ES DE T1'''
if len(data_array[1]) == 9:
  data_array = np.c_[data_array, data_array[:,2]*100/data_array[:,1]]
print('La forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))
```

```
'''(5) NUEVO ARRAY DE FORMA rs[i,j] = r[i] * s[j]'''

# Se toman los indices de los 10 mayores valores de la columna r:
max10_r_indexes = data_array[:,5].argsort()[-10:]
# En base a los índices se toman los 10 valores mayores de la columna r:
max10_r = data_array[:,5][max10_r_indexes]
# Se obtienen los valores corresondientes de a columna s:
s_corres = data_array[:,6][max10_r_indexes]
print("Los 10 valores más altos de la columna 'r' son:\n{}".format(list(max10_r)))
print("Los 10 valores correspondientes de la columna 's' son:\n{}".format(list(s_corres)))

# Creación de una matriz con los valores r[i]*s[j] y el heatmap:
matriz = np.around(np.outer(s_corres, max10_r), decimals = 2)
fig, ax = plt.subplots()
im = ax.imshow(matriz, origin='lower')

# Configuración de los textos de as etiquetas:
ax.set_xticks(np.arange(len(max10_r)))
ax.set_yticks(np.arange(len(s_corres)))
ax.set_xticklabels(max10_r)
ax.set_yticklabels(s_corres)
# Rotación de las etiquetas y cnfiguración de su alineación:
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Ciclos sobre los datos para poner los valores:
for i in range(len(max10_r)):
    for j in range(len(s_corres)):
        text = ax.text(j, i, matriz[i, j],
                       ha="center", va="center", color="w")

# Configuración del tamaño y fondo:
fig.set_size_inches(7, 7)
fig.patch.set_facecolor('xkcd:white')

# Configuración del título y del nombre de las variables:
ax.set_title("Matriz de Multiplicaciones Columna 'r' y Columna 's'")
plt.xlabel("Valores Columna 'r'")
plt.ylabel("Valores Columna 's'")
fig.tight_layout()
plt.show()
```

```
'''(6) GRÁFICO DE f(t) = r*s/t'''
plt.plot(data_array[:,0], data_array[:,7]/data_array[:,0])
plt.xlabel('Días')
plt.ylabel('Resultados')
plt.title('Gráfico de f(t) = r*s/t')
plt.show()
```

```
'''(7) CREACIÓN DE LA COLUMNA R = e**(rs/t)'''

if len(data_array[1]) == 10:
  data_array = np.c_[data_array,np.e*((data_array[:,5]*data_array[:,6])/(data_array[:,0]))]

print('La forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))
  
plt.plot(data_array[:,0], data_array[:,10])
plt.xlabel('Días')
plt.ylabel('Resultados')
plt.title('Gráfico de la Columna R')
plt.style.use('dark_background')
plt.show()
Maximo_R = data_array[:,10].max()
print(Maximo_R)
Minimo_R = data_array[:,10].min()
print(Minimo_R)
Media_R = data_array[:,10].mean()
print(Media_R)
Dev_Est_R = data_array[:,10].std()
print(Dev_Est_R)


```

```
'''(8) CREACIÓN DE LA COLUMNA W = log(rs/t)'''

if len(data_array[1]) == 11:
  data_array = np.c_[data_array, np.log(data_array[:,7]/data_array[:,0])]
  
print('La forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))

# Creación del gráfico:
plt.plot(data_array[:,0], data_array[:,11])
plt.xlabel('Días')
plt.ylabel('Resultados de W(t)=log(rs/t)')
plt.title('Gráfico de W(t)=log(rs/t)')
plt.style.use('dark_background')
plt.show()

# Cálculo de valores:
max_W = data_array[:,11].max()
min_W = data_array[:,11].min()
prom_W = data_array[:,11].mean()
dev_est__W = data_array[:,11].std()

# Imresión de valores:
print('El máximo es: {}'.format(max_W))
print('El mínimo es: {}'.format(min_W))
print('El promedio es: {}'.format(prom_W))
print('La desviación estándar es: {}'.format(dev_est__W))


```
