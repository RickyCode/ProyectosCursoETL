```
'''(1) LECTURA DE LOS DATOS Y ALMACENAMIENTO EN ARRAY DE NUMPY'''

# Pra cargar los datos
# data = rcsv('/content/drive/My Drive/Proyecto 02/datos.csv') # Compu Carlos
data = rcsv('/content/drive/My Drive/Classroom/Machine Learning C3 - 2 GRU/Ricardo Andres Del Rio Guzman - datos.csv') # Compu Ricky

# Creación del array con los datos:
print('Se están cargando los datos...\n')
data_array = np.array(data.values)
print(data_array)
print('\nLa forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))
```

```
'''(2) CREACIÓN DE UNA FUNCIÓN QUE PERMITA CALCULAR MEDIDAS DESCRIPTIVAS Y GRAFICAR:'''

def moda(columna):
  # Esta función calcula la moda de una columna dada.
  (values,counts) = np.unique(data_array[:,columna],return_counts=True)
  ind = np.argmax(counts) # Retorna el índice del valor más grande del vector counts.
  return values[ind] # Se retorna el valor correspondiente del vector values.

'''(2.a) CÁLCUO DE LAS MEDIDAS DESCRIPTIVAS'''
def calc_medidas_descriptivas(data_array, columna):
  funcs = [np.max, np.min, np.mean, np.std]
  medidas_descriptivas = np.zeros(5)
  # Se calculan los max, min, prom y devs. estandar:
  for medida in range(0,4):
    medidas_descriptivas[medida] = funcs[medida](data_array[:,columna])
  # Se calcula la desvación estándar:
  medidas_descriptivas[4] = medidas_descriptivas[3] / medidas_descriptivas[2] # Dev Est / Prom
  return medidas_descriptivas


'''(2.b) GRÁFICOS, (2,c) BOXPLOTS, (2,d) GRÁFICOS CON DATOS CORREGIDOS'''
def crear_grafico(data_array, columna, dic, n, boxplot=False):  
  if boxplot:
    axs[n].boxplot(data_array[:,columna])
  else:
    axs[n].plot(data_array[:,0],data_array[:,columna])
    axs[n].set(xlabel= dic[columna]['xlabel'])
  axs[n].set(ylabel=dic[columna]['ylabel'])
  axs[n].title.set_text(dic[columna]['title'])
  
  
''' (2,c) QUITAR OUTLIERS'''
def quitar_outliers(data_array, i, medidas_descriptivas):
  indices_arriba = np.where(data_array[:,i] > (medidas_descriptivas[2] + 3*medidas_descriptivas[3])) # Media + 3 * Dev Estandar
  for k in indices_arriba:
    data_array[:,i][k] = moda(i)
  indices_abajo = np.where(data_array[:,i] < (medidas_descriptivas[2] - 3*medidas_descriptivas[3])) # Media - 3 * Dev Estandar
  for h in indices_abajo:
    data_array[:,i][h] = moda(i)
  return data_array


def mostrar_medidas(iterable):
  lista_nom = ['Máximo:', 'Mínimo:', 'Promedio:', 'Des. Estándar:', 'Dev. Est. Relativa:']
  print()
  for i in range(len(iterable)):
    print('\t', lista_nom[i] + '{0:.2f}'.format(iterable[i]).rjust(25-len(lista_nom[i]), ' '), end='    |')
  print('\n')
    
    
  
### FUNCION PRINCIPAL:
  
  
def calculo_limpieza(data_array, dic): 
  global fig, axs
  
  # Recorremos las columnas de la matriz:
  for i in range(1,7):
    fig, axs = plt.subplots(1,3)
    print('\n\n' + '*'*160 + '\n' + dic[i]['title'].center(160, ' ').upper() + '\n' + '*'*160)
    med_desc = calc_medidas_descriptivas(data_array, i)
    mostrar_medidas(med_desc)
    crear_grafico(data_array, i, dic, 0)
    crear_grafico(data_array, i, dic, 1, True)
    data_array = quitar_outliers(data_array, i, med_desc) # Se sobrescribe el data_array   
    crear_grafico(data_array, i, dic, 2) # Gráficos corregidos
    print(dic[i]['comentario'])

    fig.set_size_inches(20, 4)
    fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                        hspace=0.4, wspace=0.3)
    fig.patch.set_facecolor('xkcd:baby blue')
    plt.show()
    
    med_desc = calc_medidas_descriptivas(data_array, i)
    mostrar_medidas(med_desc)
  
  
calculo_limpieza(data_array, dic) 
```

```
 '''(3) CREACION DE LAS COLUMNAS rs y rsacum'''
# Creación de columna rs:
if len(data_array[1]) == 7:
  data_array = np.c_[data_array, data_array[:,5]*data_array[:,6]]
  print('Se ha creado la columna rs!')
  
  # Creación de columna rsacum:
if len(data_array[1]) == 8:
  data_array = np.c_[data_array, np.cumsum(data_array[:,7])]
print('Las colunas "rs" y "rsacum" están creadas!')
print('La forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))
```

```
'''(4) CREACION DE COLUMNA CON EL PORCENTAJE QUE T2 ES DE T1'''
if len(data_array[1]) == 9:
  data_array = np.c_[data_array, data_array[:,2]*100/data_array[:,1]]
print('La columna con el pocentaje que T2 es de T1 está creada!')
print('La forma actual de la matriz es: {} filas, {} columnas'.format(data_array.shape[0], data_array.shape[1]))
```

```
'''(5) NUEVO ARRAY DE FORMA rs[i,j] = r[i] * s[j]'''

# Se toman los indices de los 10 mayores valores de la columna r:
max10_r_indexes = data_array[:,5].argsort()[-10:] # argsort() retorna una lista de índices ordenados
# En base a los índices se toman los 10 valores mayores de la columna r:
max10_r = data_array[:,5][max10_r_indexes]
# Se obtienen los valores corresondientes de a columna s:
s_corres = data_array[:,6][max10_r_indexes]
print("Los 10 valores más altos de la columna 'r' son:\n{}\n".format(list(max10_r)))
print("Los 10 valores correspondientes de la columna 's' son:\n{}\n".format(list(s_corres)))

# Creación de una matriz con los valores r[i]*s[j] y el heatmap:
matriz = np.around(np.outer(s_corres, max10_r), decimals = 2)
fig, ax = plt.subplots()
im = ax.imshow(matriz, origin='lower')

# Configuración de los textos de las etiquetas:
ax.set_xticks(np.arange(len(max10_r)))
ax.set_yticks(np.arange(len(s_corres)))
ax.set_xticklabels(max10_r)
ax.set_yticklabels(s_corres)
# Rotación de las etiquetas y cnfiguración de su alineación:
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Ciclos sobre los datos para poner los valores:
for i in range(len(max10_r)):
    for j in range(len(s_corres)):
        text = ax.text(j, i, matriz[i, j],
                       ha="center", va="center", color="w")

# Configuración del tamaño y fondo:
fig.set_size_inches(7, 7)
fig.patch.set_facecolor('xkcd:white')

# Configuración del título y del nombre de las variables:
ax.set_title("Matriz de Multiplicaciones Columna 'r' y Columna 's'")
plt.xlabel("Valores Columna 'r'")
plt.ylabel("Valores Columna 's'")
fig.tight_layout()
plt.show()
```

```
'''(6) GRÁFICO DE f(t) = r*s/t'''
funcion_f = data_array[:,7]/data_array[:,0]
fig = plt.figure(figsize=(8, 6), dpi=80, facecolor='lightblue', edgecolor='k')
plt.plot(data_array[:,0], funcion_f)
plt.xlabel('Días')
plt.ylabel('Resultados')
plt.title('Gráfico de f(t) = r*s/t')
plt.show()

# Impresión de valores:
print('\nEl máximo es: {}'.format(funcion_f.max()))
print('El mínimo es: {}'.format(funcion_f.min()))
print('El promedio es: {}'.format(funcion_f.mean()))
print('La desviación estándar es: {}'.format(funcion_f.std()))
```

```
'''(7) CREACIÓN DE LA COLUMNA R = e**(rs/t)'''

# Creacíón de la columna:
if len(data_array[1]) == 10:
  data_array = np.c_[data_array, np.e**((data_array[:,5]*data_array[:,6])/(data_array[:,0]))]

print('La columna "R" está creada!')
print('La forma actual de la matriz es: {} filas, {} columnas\n'.format(data_array.shape[0], data_array.shape[1]))

# Creación del gráfico:
fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='lightblue', edgecolor='k')
plt.plot(data_array[:,0], data_array[:,10])
plt.xlabel('Días')
plt.ylabel('Resultados')
plt.title('Gráfico de la Columna R')
plt.show()

# Cálculo de medidas descriptivas:
Maximo_R = data_array[:,10].max()
Minimo_R = data_array[:,10].min()
Media_R = data_array[:,10].mean()
Dev_Est_R = data_array[:,10].std()

# Imresión de valores:
print('\nEl máximo es: {}'.format(Maximo_R))
print('El mínimo es: {}'.format(Minimo_R))
print('El promedio es: {}'.format(Media_R))
print('La desviación estándar es: {}\n'.format(Dev_Est_R))

# Cometario:
print('''COMENTARIO
Se puede observar que la forma del gráfico es muy similar a la del gráfico de la parte 6.
La diferencia es que el valor mínimo es uno, mietras que en el gráfico anterior el mínimo es cero.
Además el rango de valores del eje vertical es mayor, pues va desde el 1 al 4, mientras que en
e anterior los valores están entre 0 y 1.4, por lo que las oscilaciones tienen mayor amplitud.''')
```

```
'''(8) CREACIÓN DE LA COLUMNA W = log(rs/t)'''

if len(data_array[1]) == 11:
  data_array = np.c_[data_array, np.log(data_array[:,7]/data_array[:,0])]

print('La columna "W" está creada!')
print('La forma actual de la matriz es: {} filas, {} columnas\n'.format(data_array.shape[0], data_array.shape[1]))

# Creación del gráfico:
fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='lightblue', edgecolor='k')
plt.plot(data_array[:,0], data_array[:,11], '-o')
plt.xlabel('Días')
plt.ylabel('Resultados de W(t)=log(rs/t)')
plt.title('Gráfico de W(t)=log(rs/t)')
plt.show()

# Cálculo de valores:
max_W = data_array[:,11].max()
min_W = data_array[:,11].min()
prom_W = data_array[:,11].mean()
dev_est__W = data_array[:,11].std()

# Impresión de valores:
print('\nEl máximo es: {}'.format(max_W))
print('El mínimo es: {}'.format(min_W))
print('El promedio es: {}'.format(prom_W))
print('La desviación estándar es: {}\n'.format(dev_est__W))

print('''COMENTARIO
Este gráfico posee grandes diferencias con el gráfico observado de la función "f(x)=rs/t".
Primero que nada este gráfico se mueve entre valores negativos llegando hasta el menos 6.
Además se observa una gran discontinuidad en este gráfico dado a una gran cantidad de valores 
que se indeterminan, cosa que no ocurre en la función. Esto lleva a ver muchos puntos separados
de los demás''')

```

```
'''(9) CALCULO Y GRAFICO DE DISTANCAS ENTRE PUNTOS'''

# Cálculo de las distancias:
x = np.sqrt((((data_array[:,1])-(data_array[:,2]))**2) + (((data_array[:,3])-(data_array[:,4]))**2))

# Creación del gráfico:
fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='lightblue', edgecolor='k')
plt.plot(data_array[:,0], x)
plt.xlabel('Días')
plt.ylabel('Distancias')
plt.title('Gráfico de la funcion D')
plt.show()

# Impresión de valores:
print('\nEl máximo es: {}'.format(x.max()))
print('El mínimo es: {}'.format(x.min()))
print('El promedio es: {}'.format(x.mean()))
print('La desviación estándar es: {}'.format(x.std()))
```
